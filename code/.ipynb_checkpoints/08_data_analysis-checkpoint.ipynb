```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
```

Leitura dados

```python
df = pd.read_csv('../data/processed/relational_df.csv', sep=';', encoding = 'utf-8',index_col=0)
```

Conjuntos de ortologos compartilhados entre as Bacterias

```python
from upsetplot import from_memberships

orthosets = df.groupby('Orthogroup').agg({'Bacteria':'unique'}).reset_index()
orthosets['Bacteria'] = orthosets['Bacteria'].apply(sorted)
orthosets
```

Decritivo geral

```python
df_count =df.groupby('Bacteria').agg(Grupos_Ortologos = ('Orthogroup','nunique')).reset_index()
barplot = sns.barplot(data=df_count, y='Bacteria',x='Grupos_Ortologos', palette='Reds')
plt.figure(figsize=(10, 6))

for p in barplot.patches:
    barplot.annotate(format(p.get_width(), '.0f'), 
                     (p.get_width(), p.get_y() + p.get_height() / 2), 
                     ha = 'left', va = 'center', 
                     size=14,
                     xytext = (5, 0), 
                     textcoords = 'offset points')

plt.savefig('../reports/BarPlot.svg', dpi=300)
plt.show()
```

```python
orthosets['n_shared'] = orthosets['Bacteria'].apply(len)
```

Verificando descriÃ§Ã£o dos genes compartilhados por ao menos 4 bacterias 

```python
shared_4 = orthosets.query('n_shared > 3').Orthogroup.unique()
```

```python
report_table = df.query('Orthogroup in @shared_4').groupby(['Orthogroup','Gene','Bacteria']).agg({'gb_accession':'first', 'description':'first'})
```

Matriz de compartilhamento de ortÃ³logos

```python
# Create a new dataframe where each row represents a pair of bacteria that share at least one ortholog
df_pairs = pd.merge(df[['Orthogroup','Bacteria']], 
                    df[['Orthogroup','Bacteria']],
                      on='Orthogroup')

# # Exclude pairs of the same bacteria
# df_pairs = df_pairs[df_pairs['Bacteria_x'] != df_pairs['Bacteria_y']]

# Create a pivot table with Bacteria_x and Bacteria_y as indices, and the count as values
pivot_table = pd.pivot_table(df_pairs, index='Bacteria_x', columns='Bacteria_y', aggfunc='size', fill_value=0)

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(pivot_table, dtype=bool))

# Mask the pivot table
masked_pivot = np.ma.array(pivot_table, mask=mask)

# Generate a heatmap with annotations
fig, ax = plt.subplots(figsize=(6, 4))
cax = ax.matshow(masked_pivot, cmap="Blues")
plt.colorbar(cax)
ax.set_title('Matriz de compartilhamento de ortÃ³logos', pad=20)

# Set x labels at the bottom with a little angle
ax.xaxis.set_ticks_position('bottom')
plt.xticks(range(len(pivot_table.columns)), pivot_table.columns, rotation=45, ha='right')
plt.yticks(range(len(pivot_table.index)), pivot_table.index)

max_val = masked_pivot.max()
for (i, j), val in np.ndenumerate(masked_pivot):
    if not mask[i, j]:
        text_color = 'white' if val > max_val / 2 else 'black'
        ax.text(j, i, f'{val:.0f}', ha='center', va='center', color=text_color)



# # Remove the top and right spines
# ax.spines['right'].set_visible(False)
# ax.spines['top'].set_visible(False)
plt.savefig('../reports/Matriz.png', dpi=300)
plt.show()
```
